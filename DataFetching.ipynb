```python
# get_yf_2years_ohlcv.py
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import os

# ----------------------
# CONFIG
# ----------------------
TICKERS = {
    "WTI": "CL=F",     # WTI Crude Oil futures
    "META": "META",    # Meta Platforms Inc.
    "URTH": "URTH"     # iShares MSCI World ETF
}

END_DATE = datetime.today()
START_DATE = END_DATE - timedelta(days=365 * 2)  # last 2 years
OUT_DIR = "data_2y_ohlcv"

os.makedirs(OUT_DIR, exist_ok=True)

# ----------------------
# DOWNLOAD & SAVE DATA
# ----------------------
all_data = []

for name, symbol in TICKERS.items():
    df = yf.download(
        symbol,
        start=START_DATE.strftime("%Y-%m-%d"),
        end=END_DATE.strftime("%Y-%m-%d"),
        interval="1d",
        auto_adjust=False,  # keep raw OHLC values
        progress=False
    )
    df = df.reset_index()
    df = df.rename(columns=str.lower)  # lowercase column names

    # Keep only required columns
    df = df[["date", "open", "high", "low", "close", "volume"]]

    # Add ticker name for identification in combined file
    df["symbol"] = name

    # Save per-ticker CSV
    df.to_csv(os.path.join(OUT_DIR, f"{name}_2y_daily.csv"), index=False)
    print(f"âœ… Saved {name} ({symbol}) data: {len(df)} rows")

    # Append to combined list
    all_data.append(df)

# ----------------------
# COMBINED FILE
# ----------------------
combined_df = pd.concat(all_data, ignore_index=True)
combined_df.to_csv(os.path.join(OUT_DIR, "All_Tickers_2y_daily.csv"), index=False)

print(f"ðŸ“‚ Combined dataset saved: {os.path.join(OUT_DIR, 'All_Tickers_2y_daily.csv')}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV files and labels (all in the same folder as this script)
files = [
    ("WTI_2y_daily.csv", "WTI"),
    ("META_2y_daily.csv", "META"),
    ("URTH_2y_daily.csv", "URTH")
]

for file, label in files:
    # Load CSV
    df = pd.read_csv(file)
    
    # Normalize column names
    df.columns = [c.lower().strip() for c in df.columns]
    
    # Handle date column variations
    if "timestamp" in df.columns and "date" not in df.columns:
        df.rename(columns={"timestamp": "date"}, inplace=True)
    
    # Handle close column variations
    if "close" not in df.columns:
        for alt in ["adj close", "adj_close", "adjclose"]:
            if alt in df.columns:
                df.rename(columns={alt: "close"}, inplace=True)
                break
    
    # Ensure correct types
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df["close"] = pd.to_numeric(df["close"], errors="coerce")
    df = df.dropna(subset=["date", "close"])
    df = df.sort_values("date")
    
    # Plot
    plt.figure(figsize=(10,5))
    plt.plot(df["date"], df["close"], label=label)
    plt.title(f"{label} â€” Close Price (Last 2 Years)")
    plt.xlabel("Date")
    plt.ylabel("Close Price")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
